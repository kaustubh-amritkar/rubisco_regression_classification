```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy
```

```python
## Reading the different performance metrics files

esm2_combined_split_performance_df = pd.read_csv("/home/kaustubh/RuBisCO_ML/regressor_models/data/regressor_output/normalized_wrt_ind_ancs/ESM-2_log_all_data_split/performance_metrics.csv")
esm2_combined_split_performance_df['split_type'] = "esm2_combined_split"

esm2_large_train_mutant_test_split_performance_df = pd.read_csv("/home/kaustubh/RuBisCO_ML/regressor_models/data/regressor_output/normalized_wrt_ind_ancs/ESM-2_log_large_mutant_data_split/performance_metrics.csv")
esm2_large_train_mutant_test_split_performance_df['split_type'] = "esm2_large-train_mutant_test"

esm2_mutant_train_large_train_split_performance_df = pd.read_csv("/home/kaustubh/RuBisCO_ML/regressor_models/data/regressor_output/normalized_wrt_ind_ancs/ESM-2_log_mutant_train_large_test/performance_metrics.csv")
esm2_mutant_train_large_train_split_performance_df['split_type'] = "esm2_mutant-train_large_test"

protT5_combined_split_performance_df = pd.read_csv("/home/kaustubh/RuBisCO_ML/regressor_models/data/regressor_output/normalized_wrt_ind_ancs/protT5_log_all_data_split/performance_metrics.csv")
protT5_combined_split_performance_df['split_type'] = "protT5_combined_split"

protT5_large_train_mutant_test_split_performance_df = pd.read_csv("/home/kaustubh/RuBisCO_ML/regressor_models/data/regressor_output/normalized_wrt_ind_ancs/protT5_log_large_mutant_data_split/performance_metrics.csv")
protT5_large_train_mutant_test_split_performance_df['split_type'] = "protT5_large-train_mutant_test"

prostT5_combined_split_performance_df = pd.read_csv("/home/kaustubh/RuBisCO_ML/regressor_models/data/regressor_output/normalized_wrt_ind_ancs/prostT5_log_all_data_split/performance_metrics.csv")
prostT5_combined_split_performance_df['split_type'] = "prostT5_combined_split"
```

```python
combined_performance_df = pd.concat([esm2_combined_split_performance_df, esm2_large_train_mutant_test_split_performance_df, esm2_mutant_train_large_train_split_performance_df, protT5_combined_split_performance_df, protT5_large_train_mutant_test_split_performance_df, prostT5_combined_split_performance_df], ignore_index=True)
combined_performance_df
```

```python
combined_performance_df[["model","spearman_rho", "split_type"]]
```

```python
## Make a plot showing the performance of the different models on the different splits

fig, ax = plt.subplots(2,2,figsize=(20, 15))
plt.subplots_adjust(top=0.9, hspace=0.3, wspace=0.3)

performance_metrics = ["spearman_rho", "r2", "mae", "rmse"]
ax_list = [ax[0][0], ax[0][1], ax[1][0], ax[1][1]]
for i, metric in enumerate(performance_metrics):
    plt.figure(figsize=(10, 6))
    df = combined_performance_df[["model", metric, "split_type"]]
    sns.scatterplot(x='model', y=metric, hue='split_type', data=df, style='split_type', s=100, ax=ax_list[i])
    ax_list[i].set_xticklabels(ax_list[i].get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```
